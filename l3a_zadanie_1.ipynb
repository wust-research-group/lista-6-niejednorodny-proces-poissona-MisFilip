```python
import numpy as np
import matplotlib.pyplot as plt
```

Cauchy

Using 'while'

```python
def cauchy_distribution(samples, loc=0, scale=1):
    cauchy_samples = np.zeros(samples)
    i = 0

    while i < samples:
        U1 = np.random.uniform(0, 1)
        U2 = np.random.uniform(-1, 1)
        W = U1**2 + U2**2
        
        if W < 1:
            X = loc + scale * U1 / U2
            cauchy_samples[i] = X
            i += 1
    
    return np.array(cauchy_samples)
```

```python
m = 10000
zeros = np.zeros(m)
cauchy_samples = cauchy_distribution(m)
plt.hist(cauchy_samples, bins=int(max(cauchy_samples)-min(cauchy_samples)), density=True, alpha=0.8, color='blue', edgecolor='black')
plt.title('Cauchy Distribution - using "while"')
plt.xlabel('Value')
plt.ylabel('Density')
plt.xlim([-20,20])
x = np.linspace(-20, 20, 1000)
y = 1 / (np.pi * (1 + x**2))
plt.plot(x, y, color='green', linestyle='--')
```

Recursive method

```python
def generator():
    x = np.random.uniform(0,1)
    y = np.random.uniform(-1, 1)
    z = x**2 + y**2
    if z < 1:
        return [x, y, z]
    else:
        return generator()
```

```python
def cauchy_dist(n, board):
    for i in range(n):
        xyz = generator()
        sample = xyz[1] / xyz[0]
        board[i] = sample
    return board
```

```python
m = 10000
zeros = np.zeros(m)
cauchy_samples = cauchy_dist(m, zeros)
plt.hist(cauchy_samples, bins=int(max(cauchy_samples)-min(cauchy_samples)), density=True, alpha=0.8, color='blue', edgecolor='black')
plt.title('Cauchy Distribution - using recursion')
plt.xlabel('Value')
plt.ylabel('Density')
plt.xlim([-20,20])
x = np.linspace(-20, 20, 1000)
y = 1 / (np.pi * (1 + x**2))
plt.plot(x, y, color='blue', linestyle='--')
```
